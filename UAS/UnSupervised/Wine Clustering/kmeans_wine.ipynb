# Train Data

```python
x = df_wine[["Alcohol", "Total_Phenols"]]  
```

# Scalling

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(x)
X_scaled = scale.transform(x)
```

# Train Model

```python
from sklearn.cluster import KMeans

kmean = KMeans(n_clusters=3)
kmean
```

# Predict Test

```python
y_cluster = kmean.fit_predict(x)
y_cluster
```

```python
df_wine['cluster'] = y_cluster
df_wine
```

```python
kmean.cluster_centers_
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=y_cluster) 

plt.scatter(kmean.cluster_centers_[:,0], kmean.cluster_centers_[:,1], marker="*", s=250, c=[0,1,2], edgecolors="k") 

plt.xlabel("Alcohol")
plt.ylabel("Total Phenols")
plt.title("KMeans k=3")
plt.show() 
```

# Elbow Method

```python
kmean.inertia_ 
```

```python
inertia = []
for i in np.arange(1,11):
    km = KMeans(n_clusters=i)
    km.fit(X_scaled)
    inertia.append(km.inertia_)
    
plt.plot(np.arange(1,11), inertia, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show() 
```
