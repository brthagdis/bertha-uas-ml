# Read Data

```python
import pandas as pd
import numpy as np
```

```python
df_pumpkin = pd.read_excel('Pumpkin_Seeds_Dataset.xlsx')
df_pumpkin
```

```python
print("Data Shape : ",df_pumpkin.shape)
```

```python
df_pumpkin.info()
```

```python
df_pumpkin.describe()
```

```python
pd.unique(df_pumpkin['Class'])
```

```python
df_pumpkin['Class'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder as label_encoder
le = label_encoder()
df_pumpkin['Class'] = le.fit_transform(df_pumpkin['Class'])
df_pumpkin['Class'].value_counts()
```

# Split Atribute

```python
x = df_pumpkin.drop('Class', axis=1)
y = df_pumpkin['Class'] 
```

# Scalling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x) 
x = scaler.transform(x)
```

# Split Train Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# Train Model

```python
from sklearn.linear_model import Perceptron
pcp= Perceptron()
pcp.fit(x_train, y_train)
pcp
```

# Predict Test

```python
y_pred = pcp.predict(x_test)
y_pred
```

# Akurasi dan Evaluasi

```python
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 

accuracy = accuracy_score(y_test, y_pred)  
report = classification_report(y_test, y_pred, target_names=le.classes_)
print("\nLaporan Klasifikasi:\n", report)
print("Akurasi Model:", accuracy)  
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

pcp_cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 7))
sns.heatmap(pcp_cm, annot=True, fmt='d', cmap='Greens', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix - PERCEPTRON Model')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=le.classes_, y=pcp_cm.diagonal(), palette="viridis")
plt.title("Perceptron Prediction Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```
