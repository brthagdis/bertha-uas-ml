# Read Data

```python
import pandas as pd
import numpy as np
```

```python
df_pumpkin = pd.read_excel('Pumpkin_Seeds_Dataset.xlsx')
df_pumpkin
```

```python
print("Data Shape : ",df_pumpkin.shape)
```

```python
df_pumpkin.info()
```

```python
df_pumpkin.describe() 
```

```python
pd.unique(df_pumpkin['Class'])
```

```python
df_pumpkin['Class'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder as label_encoder
le = label_encoder()
df_pumpkin['Class'] = le.fit_transform(df_pumpkin['Class'])
df_pumpkin['Class'].value_counts()
```

# Split Atribute

```python
x = df_pumpkin.drop('Class', axis=1)
y = df_pumpkin['Class'] 
```

# Split Train Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=0)
```

# Train Model

```python
from sklearn.svm import SVC
svm = SVC(kernel='poly').fit(x_train, y_train)
svm

#KERNEL POLY AKURASI 0.9656666666666667
#KERNEL LINEAR AKURASI 0.9253333333333333
#KERNEL RBF AKURASI 0.931
#KERNEL SIGMOID AKURASI 0.488
```

# Predict Test

```python
y_pred = svm.predict(x_test)
y_pred
```

# Akurasi dan Evaluasi

```python
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 

accuracy = accuracy_score(y_test, y_pred)  
report = classification_report(y_test, y_pred, target_names=le.classes_)
print("\nLaporan Klasifikasi:\n", report)
print("Akurasi Model:", accuracy) 
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

svm_cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(svm_cm, annot=True, fmt='d', cmap='Greens', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix - SVM Model')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))  
sns.barplot(x=le.classes_, y=svm_cm.diagonal(), palette="viridis") 
plt.title("SVM Prediction Class Distribution")  
plt.xlabel("Class")  
plt.ylabel("Count")  
plt.xticks(rotation=85)  
plt.tight_layout()  
plt.show() 
```
